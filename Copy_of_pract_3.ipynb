```python
from pandas.core.indexes.extension import Index
import pandas as pd
import numpy as np

df = pd.DataFrame({'Roll.No': [12, 18, 12,10],
                   'Sem2': [1,3,2,5]},
                   index = ['Aayush','Amitesh','Arnav','Aditya'])
print(df)
```



 **1.Min-Max Scalar**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df1= pd.DataFrame(scaler.fit_transform(df),
                  columns=['Roll.No','Sem2'],
                  index=['Aayush','Amitesh','Arnav','Aditya']) 
df1
```

**2. standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df2=pd.DataFrame(scaler.fit_transform(df),
columns =['Roll.No','Sem2'],
index=['Aayush','Amitesh','Arnav','Aditya'])
df2
```

**3.Max Abs Scaler bold text**

```python
from sklearn.preprocessing import MaxAbsScaler
scaler= MaxAbsScaler()
df3= pd.DataFrame(scaler.fit_transform(df),
                  columns=['Roll.No','Sem2'],
                  index=['Aayush','Amitesh','Arnav','Aditya'])
df3
```

**4. Robust scaler**

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
df4=pd.DataFrame(scaler.fit_transform(df),
columns= ['Roll.No','Sem2'],
index=['Aayush','Amiteh','Arnav','Aditya'])
df4
```

**5. Quantile tramsformer Scaler**

```python
from sklearn.preprocessing import QuantileTransformer
scaler= QuantileTransformer()
df5=pd.DataFrame(scaler.fit_transform(df),
columns= ['Roll.No','Sem2'],
index=['Aayush','Amiteh','Arnav','Aditya'])
df5
```
